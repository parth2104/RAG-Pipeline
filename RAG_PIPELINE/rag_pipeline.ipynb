```python
from langchain.document_loaders import TextLoader
```

```python
text_loader=TextLoader(r"D:\RAG\ml_ai_corpus.txt",encoding='utf-8')

text=text_loader.load() 
```

```python
text
```

```python
text_string=text[0].page_content
```

```python
type(text_string)
```

```python
print(f"page content {text[0].page_content[:800]}")
print(f"metadata {text[0].metadata}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_spliter=RecursiveCharacterTextSplitter(
    separators=[',','\n','\n\n',"."],
    chunk_size=200,
    chunk_overlap=20,
    length_function=len
)
```

```python
chunking=text_spliter.split_text(text_string)
```

```python
print(f"total chunks are {len(chunking[0])}")
chunking
```

```python
from langchain_openai import OpenAIEmbeddings
import os
from dotenv import load_dotenv

embed = OpenAIEmbeddings(
    api_key=os.getenv("api_key"),  
    model="text-embedding-3-small",        
    base_url=os.getenv("endpoints")  
)
```

```python
vector=embed.embed_documents(chunking)
```

```python
print(f"{vector[0]}")
print(f"{vector[1]}")
```

```python
from langchain.vectorstores import Chroma

vectors=Chroma.from_texts(
    texts=chunking,
    embedding=embed,
    persist_directory="vactor_db",
    collection_name="RAG_PIPELINE_DB"
)
```

```python
query="what is  ai and  ml also give me their uses ? "
```

```python
s=vectors.similarity_search(query)
```

```python
s
```

```python
s=vectors.similarity_search_with_score(query,k=3)
```

```python
s
```

```python
query="what are the coman models use in ml ? "
```

```python
s=vectors.similarity_search(query,k=8)
```

```python
s
```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(
    model="gpt-5-nano-2025-08-07",
    api_key=os.getenv("api_key"),
    openai_api_base= os.getenv("endpoints")
)
```

```python
llm
```

```python
llm=llm.invoke("what is deep learning ?")
```

```python
print(llm.content)
```

```python
from langchain.prompts import ChatPromptTemplate 

system_promt=''' You are an assistant for question answering.
use the following retrived context to answer the qustion.
if you do not know the answer, you say I do not know

context :{context}'''

promt=ChatPromptTemplate.from_messages([
    ('system',system_promt),
    ('human','{input}')
]

)
```

```python
promt
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

documents=create_stuff_documents_chain(llm,promt)
```

```python
documents
```

```python
retriver=vectors.as_retriever(
    search_kwarg={'k':5}
)
retriver
```

```python
from langchain.chains import create_retrieval_chain

reg=create_retrieval_chain(retriver,documents)
```

```python
reg
```

```python
response=reg.invoke({"input":"what is the AI,ML and Data science ? also give me in details."})
```

```python
response["answer"]
```
