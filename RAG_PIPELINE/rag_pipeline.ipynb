```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_spliter=RecursiveCharacterTextSplitter(
    separators=[',','\n','\n\n',"."],
    chunk_size=200,
    chunk_overlap=20,
    length_function=len
)
```

```python
chunking=text_spliter.split_text(text_string)
```

```python
print(f"total chunks are {len(chunking[0])}")
chunking
```

```python
from langchain_openai import OpenAIEmbeddings
import os
from dotenv import load_dotenv

embed = OpenAIEmbeddings(
    api_key=os.getenv("api_key"),  
    model="text-embedding-3-small",        
    base_url=os.getenv("endpoints")  
)
```

```python
vector=embed.embed_documents(chunking)
```

```python
print(f"{vector[0]}")
print(f"{vector[1]}")
```

```python
from langchain.vectorstores import Chroma

vectors=Chroma.from_texts(
    texts=chunking,
    embedding=embed,
    persist_directory="vactor_db",
    collection_name="RAG_PIPELINE_DB"
)
```

```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(
    model="gpt-5-nano-2025-08-07",
    api_key=os.getenv("api_key"),
    openai_api_base= os.getenv("endpoints")
)
```

```python
llm
```

```python
llm=llm.invoke("what is deep learning ?")
```

```python
print(llm.content)
```

```python
from langchain.prompts import ChatPromptTemplate 

system_promt=''' You are an assistant for question answering.
use the following retrived context to answer the qustion.
if you do not know the answer, you say I do not know

context :{context}'''

promt=ChatPromptTemplate.from_messages([
    ('system',system_promt),
    ('human','{input}')
]

)
```

```python
promt
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

documents=create_stuff_documents_chain(llm,promt)
```

```python
documents
```

```python
retriver=vectors.as_retriever(
    search_kwarg={'k':5}
)
retriver
```

```python
from langchain.chains import create_retrieval_chain

reg=create_retrieval_chain(retriver,documents)
```

```python
reg
```

```python
response=reg.invoke({"input":"what is the AI,ML and Data science ? also give me in details."})
```

```python
response["answer"]
```

```python
# Conversational chat retrival

from langchain.chains import create_history_aware_retriever
from langchain.prompts import MessagesPlaceholder 
from langchain_core.messages import AIMessage,HumanMessage
```

```python
conversational_chat=''' given a chat history and latest user question with might refrence context 
in chat history ,formulate the standalone question which can be understood without the chat history 
,Do not answer the question just reformulate it if neede  otherwise return it as it is   '''

conver_prompt=ChatPromptTemplate.from_messages(
    [("system",conversational_chat),
     MessagesPlaceholder("chat_history"),
     ("human","{input}")]
)
```

```python
# create history retriver 
history_retriver=create_history_aware_retriever(
    llm,reg,conver_prompt
)
```

```python
history_retriver
print("history_retriver")
```

```python
qa_system='''you are an assistant for quuestion answering tasks
use the following pleces of retrevied context to answer the question . if you do  not the naswer 
just say that you do not know. 

context :{context}'''

qa_prompt=ChatPromptTemplate.from_messages([
    ("system",qa_system),
    MessagesPlaceholder("chat_history"),
    ("human","{input}")
])

qa_chain=create_stuff_documents_chain(llm,qa_prompt)
```

```python
# create conversational chain 

history_chain=create_retrieval_chain(
    history_retriver,
    qa_chain
)
```

```python
history_chain
print("history_chain")
```

```python
chat_history=[]


answer_one = history_chain.invoke({
    "chat_history": chat_history,
    "input": "what is machine learning?"
})

print("Q1 what is machine learning?")
print(f"{answer_one}")
```
