```python
import os, requests
from dotenv import load_dotenv

load_dotenv()
endpoint = os.getenv("endpoints")
api = os.getenv("api_key") 

response = requests.post(
    endpoint,
    headers={"Authorization": f"Bearer {api}"},
    json={
        "model": "gpt-4.1-nano", 
        "messages": [{"role": "user", "content": "Hello"}]  
    }
)
```
