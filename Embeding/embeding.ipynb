```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2")
embedding

```

```python
text="Hello , my name is parth and I am Learnning RAG"

embeddings=embedding.embed_query(text)
print(f" text : {text}")
print(f" Embeding : {embeddings}")
print(f"lenght of embeding :{len(embeddings)}")
```

```python
sentence=[" love drinking coffee in the morning.",
"The weather is very pleasant today.",
"She went to the supermarket to buy vegetables.",
"He is learning how to play the guitar.",
"The children are playing outside in the park."]

embed=embedding.embed_documents(sentence)
```

```python
print(embed)
```
